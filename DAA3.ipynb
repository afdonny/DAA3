```python
class Stack:
  def __init__(self):
    self.items = []
  def isEmpty(self):
    return self.items==[]
  def push(self,item):
    self.items.append(item)
  def pop(self):
    return self.items[len(self.items)-1]
  def size(self):
    return len(self.item)
```

```python
stack=Stack()
stack.push('Red')
stack.push('Green')
stack.push('Blue')
stack.push('Yellow')
```

```python
class Queue(object):
  def __init__(self):
    self.items=[]
  def isEmpty(self):
      return self.items == []
  def enqueue(self,item):
      self.items.insert(0,item)
  def dequeue(self):
      return self.items.pop()
  def size(self):
      return len(self.items)
queue=Queue()
queue.enqueue('Red')
queue.enqueue('Green')
queue.enqueue('Blue')
queue.enqueue('Yellow')
print(queue.size())
print(queue.dequeue())
print(queue.dequeue())
```
